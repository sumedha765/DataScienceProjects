```python
df_train['Day_of_week']=df_train['Day_No'].apply(lambda x:x%7)
df_test['Day_of_week']=df_test['Day_No'].apply(lambda x:x%7)

train1=pd.get_dummies(df_train,columns=['Course_Domain','Course_Type'])
test1=pd.get_dummies(df_test,columns=['Course_Domain','Course_Type'])

print(train1)
print(test1)

df_train = pd.merge(df_train, train1, how="outer")
df_test = pd.merge(df_test, test1, how="outer")

print(df_train.head())
print(df_test.head())
```

```python
X = df_train.drop(['Sales','Day_No', 'ID', 'User_Traffic','Course_Domain','Course_Type'], axis=1)
Y = df_train["Sales"]

df_test = df_test.drop(['ID','Day_No','Course_Domain','Course_Type'], axis=1)
```

```python
# Data is divided into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=97)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Feature Selection technique:SelectKBest.
select_feature = SelectKBest(chi2,k=10)
fit = select_feature.fit(X_train, y_train)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Features','Score']
print(featureScores.nlargest(10,'Score'))
x_train_new = select_feature.transform(X_train)
x_test_new = select_feature.transform(X_test)
df_test_new = select_feature.transform(df_test)

# # # Feature Scaling:
scaler = MinMaxScaler()
scaler.fit(x_train_new)
x_train_new = scaler.transform(x_train_new)
x_test_new = scaler.transform(x_test_new)
print(x_train_new)
print(x_test_new)
df_test_new = scaler.transform(df_test_new)
```

```python
# KNN as Data is not high dimensional.
classifier = KNeighborsRegressor(n_neighbors=9,p=1)
training = time()
classifier.fit(x_train_new, y_train)
print("Training time for KNN:", round(time() - training, 3), "s")
testing = time()
y_pred = classifier.predict(x_test_new)
print("Prediction time for KNN:", round(time() - testing, 3), "s")
print(1000 * math.sqrt(mean_squared_log_error(y_test, y_pred)))

def rmsle1000(y_true, y_pred):
    return np.sqrt(np.mean(np.power(np.log1p(y_true + 1) - np.log1p(y_pred + 1), 2))) *1000

classifier = xgboost.XGBRegressor(random_state=10)
classifier.fit(x_train_new,y_train)
y_pred= classifier.predict(x_test_new)
print(rmsle1000(y_test, y_pred))

classifier = RandomForestRegressor(random_state=10)
classifier.fit(x_train_new,y_train)
y_pred= classifier.predict(x_test_new)
print(1000 * math.sqrt(mean_squared_log_error(y_test, y_pred)))

final_submission = classifier.predict(df_test_new)
final_submission = pd.DataFrame(final_submission)
final_submission.to_csv("sample_submission.csv", index=False)
```
